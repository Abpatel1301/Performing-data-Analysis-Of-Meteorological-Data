```python
df=df.drop([ 'Summary', 'Precip Type', 'Temperature (C)', 'Wind Speed (km/h)',
       'Wind Bearing (degrees)', 'Visibility (km)', 'Pressure (millibars)',
       'Daily Summary'],axis=1)
df
```

```python
df['Formatted Date']=pd.to_datetime(df['Formatted Date'],utc=True)
```

```python
df=df.set_index('Formatted Date')
df=df[[ 'Apparent Temperature (C)', 'Humidity']].resample('M').mean()
df
```

```python

plt.plot(df['Humidity'],color='g',label='Humidity')
plt.plot(df['Apparent Temperature (C)'],color='r',label='Apparent Temperature (C)')

plt.xlabel('years')
plt.ylabel('measuerment')
plt.title('Apparent Temperature (C) vs humidity')

plt.legend(loc='best')

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.show()
```

```python
january = df[df.index.month==1]
plt.plot(january.loc['2006-01-31' : '2016-01-31','Apparent Temperature (C)'],label='Apparent Temperature (C)')
plt.plot(january.loc['2006-01-31' : '2016-01-31','Humidity'],label='Humidity')

plt.xlabel('years')
plt.ylabel('measuerment')
plt.title('Apparent Temperature (C) vs humidity In the month of January')


plt.legend(loc='best')
plt.show()
```
